```python
from matplotlib import pyplot as plt
import numpy as np
from qiskit import *
from qiskit.visualization import plot_bloch_vector
```

```python
plt.figure()
#get the current axes on the current figure
ax = plt.gca()
#plot a 2d field of arrows
#arrow points from (x,y) to (x+u, y+v)
ax.quiver([3], [5], angles='xy', scale_units='xy', scale=1)
ax.set_xlim([-1, 10])
ax.set_ylim([-1, 10])
plt.draw()
plt.show()
```

## Definitions ##

**Hermitian matrix**: any matrix that is equal to its own conjugate transpose (dagger)

**Unitary matrix**: any matrix that has the property *M*<sup>-1</sup>*M* = *I* or *M* *M*<sup>-1</sup> = *I*, where *I* is the identity matrix
