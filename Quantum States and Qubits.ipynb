```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
%config InlineBackend.figure_format = 'svg' 
```

```python
n = 8
#number of qubits
n_qb = n
#number of output bits (classical bits)
n_cb = n
qc_output = QuantumCircuit(n_qb, n_cb)

#add a measurement to the circuit for each qubit
for i in range(n):
    qc_output.measure(i, i)
```

```python
qc_output.draw(output="mpl")
```

```python
counts = execute(qc_output, Aer.get_backend("qasm_simulator")).result().get_counts()
plot_histogram(counts)
```

```python
#circuit with n qubits
qc_encode = QuantumCircuit(n)
#apply a not gate to the 7th qubit
qc_encode.x(7)
#see the circuit diagram
qc_encode.draw(output="mpl")
```

```python
qc = qc_encode + qc_output
```

```python
qc.draw(output="mpl")
```

```python
counts = execute(qc, Aer.get_backend("qasm_simulator")).result().get_counts()
plot_histogram(counts)
```

```python
qc_encode_age = QuantumCircuit(n)
qc_encode_age.x(3)
qc_encode_age.x(4)

qc_age = qc_encode_age + qc_output
qc_age.draw(output="mpl")
```

```python
counts = execute(qc_age, Aer.get_backend("qasm_simulator")).result().get_counts()
plot_histogram(counts)
```
